### 4.1.1 Distribution of All fast food restaurants resides in U.S According to State.

```python
def group_lower_ranking_values(column):                                      #User defined function for grouping the states and sum values of states
    pie_counts = fast_food_Rest.groupby(column).agg('count')
    pct_value = pie_counts[lambda df: df.columns[0]].quantile(.75)
    values_below_pct_value = pie_counts[lambda df: df.columns[0]].loc[lambda s: s < pct_value].index.values
    def fix_values(row):                                                    #group rows into other for combination of states whose value less than pct_value      
        if row[column] in values_below_pct_value:
            row[column] = 'Other'
        return row 
    pie_grouped = fast_food_Rest.apply(fix_values, axis=1).groupby(column).agg('count')
    return pie_grouped

pie_state = group_lower_ranking_values('province')
pie_labels =pie_state.name.sort_values().index
pie_counts = pie_state.name.sort_values()
plt.figure(1, figsize=(20,10))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
plt.subplot(the_grid[0, 1], aspect=1, title='State vise Distribution of vendor')
source_pie = plt.pie(pie_counts, labels=pie_labels, autopct='%1.1f%%', shadow=True, colors=colors)
```

### Observation: 

- __CA__ state have __6.8%__ fast food restaurants which is higher than remaining state resides in __U.S__.

```python
def group_lower_ranking_values(column):                                      #User defined function for grouping the states and sum values of states
    pie_counts = fast_food_rest_bur_king.groupby(column).agg('count')
    pct_value = pie_counts[lambda df: df.columns[0]].quantile(.75)
    values_below_pct_value = pie_counts[lambda df: df.columns[0]].loc[lambda s: s < pct_value].index.values
    def fix_values(row):                                                    #group rows into other for combination of states whose value less than pct_value      
        if row[column] in values_below_pct_value:
            row[column] = 'Other'
        return row 
    pie_grouped = fast_food_rest_bur_king.apply(fix_values, axis=1).groupby(column).agg('count')
    return pie_grouped

pie_state = group_lower_ranking_values('province')
pie_labels =pie_state.name.sort_values().index
pie_counts = pie_state.name.sort_values()
plt.figure(1, figsize=(20,10))
the_grid = GridSpec(2, 2)


cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
plt.subplot(the_grid[0, 1], aspect=1, title='burger king Restaurants Distribution Across United state')
source_pie = plt.pie(pie_counts, labels=pie_labels, autopct='%1.1f%%', shadow=True, colors=colors)
```
